```python
import pickle

metrics = list()

for idx in range(2, 13):
    metrics.append(pickle.load(open(f'results/transformerwithoutdenoising/metrics/metrics_{idx}.p', 'rb')))
```

```python
agent1_f1_score = list()
agent2_entropy = list()
for idx in range(11):
    agent1_f1_score.append(metrics[idx]['agent1_f1score'])
    agent2_f1_score.append(metrics[idx]['agent2_f1score'])
```

```python
agent1_entropy = list()
mutual_information = list()
for idx in range(11):
    agent1_entropy.append(metrics[idx]['agent1_entropy'])
    mutual_information.append(metrics[idx]['mutual_information'])

plt.figure(figsize=(6,6))
for idx in range(11):
    sns.lineplot(x=range(2, 13),y=agent1_entropy)
    sns.lineplot(x=range(2, 13),y=mutual_information)
plt.xlabel('Number of bits')
plt.ylabel('Agent 1')
plt.legend(['Entropy', 'Mutual Information'])
```

```python
agent2_entropy = list()
mutual_information = list()
for idx in range(11):
    agent2_entropy.append(metrics[idx]['agent2_entropy'])
    mutual_information.append(metrics[idx]['mutual_information'])

plt.figure(figsize=(6,6))
for idx in range(11):
    sns.lineplot(x=range(2, 13),y=agent2_entropy)
    sns.lineplot(x=range(2, 13),y=mutual_information)
plt.xlabel('Number of symbols')
plt.ylabel('Number of bits')
plt.legend(['Entropy', 'Mutual Information'])
```

```python
plt.figure(figsize=(6,6))
agent1_uniqueness = list()
agent2_uniqueness = list()
for idx in range(11):
    agent1_uniqueness.append(metrics[idx]['agent1_uniqueness'])
    agent2_uniqueness.append(metrics[idx]['agent2_uniqueness'])

sns.lineplot(x=range(2, 13),y=agent1_uniqueness)
sns.lineplot(x=range(2, 13),y=agent2_uniqueness)
plt.xlabel('Number of Symbols')
plt.ylabel('Uniqueness')
plt.legend(['Agent 1','Agent 2'])
```

## Conclusion

The generated image of the auto encoders are very clear even though the messages are far apart.  The auto encoder seems to favor smallar entropy. Mutual Information is quite good and it shows in the image generated using the symbols generated by another auto encoder.

It seems the transformer-based auto encoders are able to understand each other message even though the messages are far apart.

The SSIM and PSNR of the generated images are better than the linear and convolution-based auto encoders
