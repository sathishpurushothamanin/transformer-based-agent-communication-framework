How far are the two sets of codes learned by the auto encoders?  We calculate the Levenshstein distance between the message generated by each auto encoder.

```python
data = dict({1: 34, 0: 19, 2: 31, 3: 11, 4: 4, 5: 1})
for k in sorted(data):
    print('Distance ', k, ' Frequency ', data[k])
```

Over 90% of the codes learned for each digits are within a edit distance of 3. The average entropy of each auto encoder is approx. 3.0. The mutual information between them is 2.8
