```python
import pickle
communication_pathway_0_loss = pickle.load(open(f'results/transformerwithoutdenoising/loss_dict_pathway_0.p', 'rb'))
communication_pathway_1_loss = pickle.load(open(f'results/transformerwithoutdenoising/loss_dict_pathway_1.p', 'rb'))
communication_pathway_2_loss = pickle.load(open(f'results/transformerwithoutdenoising/loss_dict_pathway_2.p', 'rb'))
communication_pathway_3_loss = pickle.load(open(f'results/transformerwithoutdenoising/loss_dict_pathway_3.p', 'rb'))
```

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

mpl.rcParams['axes.titlesize'] = 24
mpl.rcParams['axes.labelsize'] = 20
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['xtick.labelsize'] = 16
mpl.rcParams['ytick.labelsize'] = 16
mpl.rcParams['legend.fontsize'] = 'Large'
mpl.rcParams['figure.figsize'] = (6,6)
mpl.rcParams['figure.dpi'] = 300
```

```python
plt.figure(figsize=(6,6))
sns.lineplot(x=range(200),y=communication_pathway_0_loss['agent1_reconstruction_loss'])
sns.lineplot(x=range(200),y=communication_pathway_1_loss['agent1_reconstruction_loss'])
sns.lineplot(x=range(200),y=communication_pathway_2_loss['agent1_reconstruction_loss'])
sns.lineplot(x=range(200),y=communication_pathway_3_loss['agent1_reconstruction_loss'])
plt.xlabel('Epoch')
plt.ylabel('Agent 1 Reconstruction Loss')
plt.legend(['No Communication', 'Shared Reconstruction', 'Shared Message', 'Shared Reconstruction and Message'])
```

## Conclusion

The generated image of the auto encoders are very clear even though the messages are far apart.  The auto encoder seems to favor smallar entropy. Mutual Information is quite good and it shows in the image generated using the symbols generated by another auto encoder.

It seems the transformer-based auto encoders are able to understand each other message even though the messages are far apart.

The SSIM and PSNR of the generated images are better than the linear and convolution-based auto encoders
